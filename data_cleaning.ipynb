```python
import pandas as pd

# Assuming you have a DataFrame named 'df' with an 'Age' column

# Define the age groups and their corresponding labels
age_groups = {
    (0, 25): 'Young',
    (26, 50): 'Middle-aged',
    (51, 100): 'Senior'
}

# Function to assign age groups based on age
def categorize_age(age):
    for age_range, group in age_groups.items():
        if age_range[0] <= age <= age_range[1]:
            return group
    return 'Unknown'

# Apply the categorization function to the 'Age' column and create a new column 'Age_Group'
df['Age_Group'] = df['Age'].apply(categorize_age)

# Print the first few rows of the DataFrame to see the results
print(df[['Age', 'Age_Group']].head())
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'df' with an 'Age_Group' column

# Calculate the frequency of each age group
age_group_counts = df['Age_Group'].value_counts()

# Calculate the percentage of each age group
age_group_percentages = (age_group_counts / len(df)) * 100

# Create a DataFrame to display the results
age_group_distribution = pd.DataFrame({'Frequency': age_group_counts, 'Percentage': age_group_percentages})

# Sort the DataFrame by age group labels (optional)
age_group_distribution = age_group_distribution.sort_index()

# Print the age group distribution
print(age_group_distribution)
```
